```python
from transformers import AutoProcessor, Gemma3ForConditionalGeneration
from PIL import Image
import torch
import requests

model_id = "google/gemma-3-4b-it"
device = "cuda" if torch.cuda.is_available() else "cpu"

model = Gemma3ForConditionalGeneration.from_pretrained(model_id, device_map="auto").eval()
processor = AutoProcessor.from_pretrained(model_id)

# Load image
image = Image.open(requests.get("http://images.cocodataset.org/val2017/000000039769.jpg", stream=True).raw)

# Construct chat-style input with *no text*
messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": ""},
            {"type": "image", "image": image}
        ]
    }
]

# Process input
inputs = processor.apply_chat_template(
    messages, add_generation_prompt=True, tokenize=True,
    return_dict=True, return_tensors="pt"
).to(model.device)

# Generate
with torch.inference_mode():
    output = model.generate(**inputs, max_new_tokens=100)

# Decode
result = processor.decode(output[0], skip_special_tokens=True)
print(result)
```

```python
from transformers import AutoProcessor, Gemma3ForConditionalGeneration
from PIL import Image
import torch
import requests

# Load model and processor
model_id = "google/gemma-3-4b-it"
model = Gemma3ForConditionalGeneration.from_pretrained(model_id, device_map="auto")
processor = AutoProcessor.from_pretrained(model_id)

# Load image
image = Image.open(requests.get("http://images.cocodataset.org/val2017/000000039769.jpg", stream=True).raw)

# Dummy chat template input with image
messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": " "},  # text is technically required
            {"type": "image", "image": image}
        ]
    }
]

# Tokenize using chat template
inputs = processor.apply_chat_template(
    messages,
    add_generation_prompt=True,
    tokenize=True,
    return_dict=True,
    return_tensors="pt"
).to(model.device)

# Access just pixel values
pixel_values = inputs["pixel_values"]

# Get image embeddings from the vision tower manually
with torch.no_grad():
    image_embeds = model.vision_tower(pixel_values)  # shape: [batch_size, num_tokens, hidden_dim]

# Extract last hidden states
image_hidden_states = image_embeds.last_hidden_state  # shape: [1, seq_len, hidden_dim]

# Pool over the sequence (tokens)
pooled = image_hidden_states.mean(dim=1)  # shape: [1, hidden_dim]

print("Image embedding shape:", pooled.shape)

```
